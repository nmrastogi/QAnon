# Karma
- We are defining Karma as a sum of score on all the posts and score on all the comments. 

```python
c=df2.groupby("author",as_index=False)["votes"].sum()
c
```

```python
c['votes'] = c['votes'].astype(float)
c
```

```python
d=df1.groupby("author",as_index=False)["votes"].sum()
d
```

```python
e=pd.merge(c,d,on="author",how="outer",suffixes=("_c","_d"))
e=e.fillna(0.0)
e["karma"]=e["votes_c"]+e["votes_d"]
e
```
