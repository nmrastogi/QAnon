```python
import numpy as np

def gini_coefficient(values):
    # Sort the values
    sorted_values = np.sort(values)
    
    # Calculate the cumulative sum of the sorted values
    cumulative_values = np.cumsum(sorted_values)
    
    # Calculate the Gini coefficient
    n = len(values)
    cumulative_sum = sum((i + 1) * sorted_values[i] for i in range(n))
    total_sum = sum(sorted_values)
    gini_index = (2 * cumulative_sum - (n + 1) * total_sum) / (n * total_sum)
    
    return gini_index

# Calculate the Gini coefficient for the unique_year_month_counts
gini_index = gini_coefficient(df['#posts'])

# Display the result
print(f"Gini Coefficient: {gini_index}")
```

```python
import numpy as np

#define function to calculate Gini coefficient
def gini(x):
    total = 0
    for i, xi in enumerate(x[:-1], 1):
        total += np.sum(np.abs(xi - x[i:]))
    return total / (len(x)**2 * np.mean(x))

gini_index = gini_coefficient(df['#posts'])

# Display the result
print(f"Gini Coefficient: {gini_index}")
```

```python
df_sorted = df.sort_values(by='#posts', ascending=False)

# Calculate the cumulative sum
df_sorted['cumulative_sum'] = df_sorted['#posts'].cumsum()

# Calculate the total sum
total_sum = df_sorted['#posts'].sum()

# Calculate the 80% value of the total sum
eighty_percent_value = 0.8 * total_sum

# Determine the cutoff for 20% of the items
cutoff_index = int(len(df_sorted) * 0.2)
cutoff_index = max(cutoff_index, 1)  # Ensure at least one item is included

# Calculate the cumulative sum for the top 20% of items
top_20_percent_value = df_sorted.iloc[cutoff_index-1]['cumulative_sum']

# Check if 80% of the total sum is within the top 20% of the items
is_80_percent_covered = top_20_percent_value >= eighty_percent_value

# Display the result
print(f"80% of the work is done by 20% of the people: {is_80_percent_covered}")
print(f"80% value: {eighty_percent_value}")
print(f"20% value: {top_20_percent_value}")
```

# Hoover Index

- https://www.un.org/en/development/desa/policy/wess/wess_dev_issues/dsp_policy_02.pdf

## Twitter

```python
import numpy as np

def hoover_index(incomes):
    """
    Calculate the Hoover index of an income distribution.
    """
    incomes = np.array(incomes)
    mean_income = np.mean(incomes)
    hoover = 0.5 * np.sum(np.abs(incomes - mean_income)) / np.sum(incomes)
    return hoover

# Example data


# Calculate Hoover index
hoover = hoover_index(df["followers_count"])
print(f'Hoover index: {hoover:.4f}')
```
