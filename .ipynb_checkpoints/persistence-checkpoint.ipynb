```python
import pandas as pd
import multiprocessing as mp

def process_chunk(chunk):
    # Convert 'createdAtformatted' to datetime
    chunk['createdAtformatted'] = pd.to_datetime(chunk['createdAtformatted'], utc=True)
    # Extract year and week from the 'createdAtformatted' column
    chunk['year_week'] = chunk['createdAtformatted'].dt.to_period('W')
    # Convert 'year_week' to string
    chunk['year_week'] = chunk['year_week'].astype(str)
    return chunk

def parallel_process(df, n_cores):
    # Split DataFrame into chunks
    chunk_size = len(df) // n_cores
    chunks = [df[i:i + chunk_size].copy() for i in range(0, len(df), chunk_size)]

    # Create a pool of workers and process each chunk in parallel
    with mp.Pool(n_cores) as pool:
        processed_chunks = pool.map(process_chunk, chunks)

    # Combine processed chunks back into a single DataFrame
    combined_df = pd.concat(processed_chunks)

    return combined_df

# Assuming your DataFrame is already loaded and named 'df'
n_cores = mp.cpu_count()  # Get the number of available CPU cores

# Process the entire DataFrame in parallel to add the 'year_week' column and convert it to string
df_with_year_week = parallel_process(df, n_cores)

# Display the first few rows of the updated DataFrame
print(df_with_year_week.head())
```

```python
import pandas as pd
import re

# Function to remove emojis from text and check if the content has only emojis
def is_emoji_only(text):
    # Ensure the input is a string
    if isinstance(text, str):
        # Regex pattern to identify emojis
        emoji_pattern = re.compile(
            "["
            u"\U0001F600-\U0001F64F"  # emoticons
            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
            u"\U0001F680-\U0001F6FF"  # transport & map symbols
            u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
            u"\U00002702-\U000027B0"  # other symbols
            u"\U000024C2-\U0001F251"  # enclosed characters
            u"\U000E0067-\U000E007F"  # special characters (such as '\U000E0067...')
            "]+", flags=re.UNICODE
        )

        # Check if the entire string consists only of emojis
        return bool(emoji_pattern.fullmatch(text.strip()))
    
    return False  # Return False for non-string types

# Assuming df is your already loaded dataframe
# Ensure the 'body' column is string type (converts all to string)
df['body'] = df['body'].astype(str)

# Define chunk size
chunksize = 10000000
total_rows = len(df)
processed_chunks = []

# Process the dataframe in chunks of 10 million rows
for i in range(0, total_rows, chunksize):
    # Get the current chunk
    chunk = df.iloc[i:i + chunksize]
    
    # Apply the emoji check function and filter out rows that contain only emojis
    chunk_cleaned = chunk[~chunk['body'].apply(is_emoji_only)]

    # Strip leading/trailing whitespaces and remove empty rows
    chunk_cleaned['body'] = chunk_cleaned['body'].str.strip()
    chunk_cleaned = chunk_cleaned[chunk_cleaned['body'] != '']  # Drop rows where 'body' is empty after stripping
    
    # Append the cleaned chunk to the list
    processed_chunks.append(chunk_cleaned)
    
    # Print progress
    print(f"Processed chunk {i // chunksize + 1} with {len(chunk_cleaned)} rows.")

# Concatenate all cleaned chunks into a single dataframe
df_cleaned = pd.concat(processed_chunks, ignore_index=True)

# Show the cleaned dataframe (or save it to a file)
print(df_cleaned.head())

# Optionally save the final cleaned dataframe to a new CSV file
# df_cleaned.to_csv('cleaned_parler_comments.csv', index=False)
```
