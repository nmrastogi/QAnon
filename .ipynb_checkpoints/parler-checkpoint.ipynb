# Sanity check code

```python
import pandas as pd

# The value we are looking for
search_value = '0000ee9f125d4efd838379ecf59f4f0d'  # index 1 of users.

# Specific columns to check plus the 'datatype' column for filtering
columns_to_check = ['post', 'datatype']  # Include 'datatype' to filter by it

# Initialize a counter for occurrences
search_value_count = 0

# Loop through the files
for i in range(0,91):  # Includes 0 to 90
    file_path = f"../largedata/parler/posts/p{i}.parquet"
    print(i)
    # Read the DataFrame with only necessary columns
    df = pd.read_parquet(file_path, columns=columns_to_check)
    
    # Filter the DataFrame for rows where datatype is 'posts'
    df_posts = df[df['datatype'] == 'comments']
    
    # Count occurrences of the search value in the 'creator' column within filtered rows
    search_value_count += (df_posts['post'].astype(str) == search_value).sum()

# Print the total number of occurrences found after all files have been processed
print(f"Total occurrences of '{search_value}' where datatype is 'comments': {search_value_count}")
```

```python
import pandas as pd
import multiprocessing as mp

def process_chunk(chunk):
    # Convert 'createdAtformatted' to datetime
    chunk['createdAtformatted'] = pd.to_datetime(chunk['createdAtformatted'], utc=True)
    # Extract year and week from the 'createdAtformatted' column
    chunk['year_week'] = chunk['createdAtformatted'].dt.to_period('W')
    # Convert 'year_week' to string
    chunk['year_week'] = chunk['year_week'].astype(str)
    return chunk

def parallel_process(df, n_cores):
    # Split DataFrame into chunks
    chunk_size = len(df) // n_cores
    chunks = [df[i:i + chunk_size].copy() for i in range(0, len(df), chunk_size)]

    # Create a pool of workers and process each chunk in parallel
    with mp.Pool(n_cores) as pool:
        processed_chunks = pool.map(process_chunk, chunks)

    # Combine processed chunks back into a single DataFrame
    combined_df = pd.concat(processed_chunks)

    return combined_df

# Assuming your DataFrame is already loaded and named 'df'
n_cores = mp.cpu_count()  # Get the number of available CPU cores

# Process the entire DataFrame in parallel to add the 'year_week' column and convert it to string
df_with_year_week = parallel_process(df, n_cores)

# Display the first few rows of the updated DataFrame
print(df_with_year_week.head())
```
