```python
import numpy as np
import statsmodels.api as sm

# Given values

n1 = 1715006
n2 = 10789022
# Calculate number of successes

x1 = 373025
x2 = 172450
# Calculate proportions
p1 = (x1 / n1) * 100
p2 = (x2 / n2) * 100

# Number of successes in each group
count = np.array([x1, x2])

# Number of observations in each group
nobs = np.array([n1, n2])

# Perform the 2-proportion z-test
stat, pval = sm.stats.proportions_ztest(count, nobs)

print(f'Statistic: {stat:.4f}, p-value: {pval:.4f}')
```

## Bonferroni correction

```python
import numpy as np
import statsmodels.api as sm

# Define the groups
groups = [
    (6163, 48094),
    (172450, 10789022),
    (373025, 1715006 )
]

# Perform pairwise comparisons
pairwise_results = []
alpha = 0.05
num_tests = 3
adjusted_alpha = alpha / num_tests

for i in range(len(groups)):
    for j in range(i + 1, len(groups)):
        x1, n1 = groups[i]
        x2, n2 = groups[j]
        
        count = np.array([x1, x2])
        nobs = np.array([n1, n2])
        
        stat, pval = sm.stats.proportions_ztest(count, nobs)
        pairwise_results.append((f'Group {i+1} vs Group {j+1}', stat, pval, pval < adjusted_alpha))

# Print results
for result in pairwise_results:
    comparison, stat, pval, significant = result
    print(f'{comparison} - Statistic: {stat:.4f}, p-value: {pval:.4f}, Significant after Bonferroni correction: {significant}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Corrected Data
data_corrected = {
    'broadcaster': [12.81, 1.59, 21.75],
    'common_user': [0.11, 10.41, 77.76],
    'hidden_influentials': [0.24, 3.01, 0.00],
    'influentials': [31.85, 0.40, 0.01],
    'lurkers': [54.96, 84.56, 0.46]
}

roles = list(data_corrected.keys())
platforms = ['dotwin', 'parler', 'twitter']

# Transpose the data so each role's values for all platforms are together
data_transposed = np.array(list(data_corrected.values())).T

# Position of bars on the x-axis
bar_width = 0.3  # Width of each bar
index = np.arange(len(roles))  # The x locations for the groups

# Create a figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Plot each platform as a group of bars
for i, platform in enumerate(platforms):
    plt.bar(index + i * bar_width, data_transposed[i], bar_width, label=platform)

# Add some text for labels, title, and custom x-axis tick labels, etc.
plt.xlabel('User Roles')
plt.ylabel('Percentage')
plt.title('Comparison of User Roles Across Platforms')
plt.xticks(index + bar_width * (len(platforms) - 1) / 2, roles)
plt.ylim(0, 100)
plt.legend(title='Platform')

# Adding percentage labels on top of each bar
for i, platform in enumerate(platforms):
    for j, v in enumerate(data_transposed[i]):
        plt.text(j + i * bar_width, v + 1, f'{v:.2f}%', ha='center', color='black', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Corrected Data
data_corrected = {
    'broadcaster': [13.53, 4.00, 11.67],
    'common_user': [0.12, 30.44, 36.28],
    'hidden_influentials': [0.25, 9.67, 38.27],
    'influentials': [33.74, 1.14, 13.55],
    'lurkers': [52.67, 54.73, 0.20]
}

roles = list(data_corrected.keys())
platforms = ['dotwin', 'parler', 'twitter']

# Transpose the data so each role's values for all platforms are together
data_transposed = np.array(list(data_corrected.values())).T

# Adjust bar width to reduce distance between plots
bar_width = 0.3  # Decrease bar width to reduce space between bars
index = np.arange(len(roles))  # The x locations for the groups

# Define a colorblind-friendly palette
colors = ['#E69F00', '#56B4E9', '#009E73']

# Create a figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Plot each platform as a group of bars
for i, (platform, color) in enumerate(zip(platforms, colors)):
    plt.bar(index + i * bar_width, data_transposed[i], bar_width, label=platform, color=color)

# Add some text for labels, title, and custom x-axis tick labels, etc.
plt.xlabel('User Roles')
plt.ylabel('Percentage')
plt.xticks(index + bar_width * (len(platforms) - 1) / 2, roles)
plt.ylim(0, 60)  # Limit y-axis to 60


# Removing top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Adding percentage labels on top of each bar
for i, platform in enumerate(platforms):
    for j, v in enumerate(data_transposed[i]):
        plt.text(j + i * bar_width, v + 1, f'{v:.2f}%', ha='center', color='black', fontsize=9)

plt.tight_layout()
plt.show()
```

## Persistence

```python
import matplotlib.pyplot as plt
import numpy as np

colors = ['#E69F00', '#56B4E9', '#009E73']
# Corrected Data
data_corrected = {
    'minimal': [95.92, 89.27, 25.06],
    'low': [3.95, 7.72, 23.25],
    'medium': [0.07, 1.62, 16.05],
    'high': [0.04, 1.36, 35.62]
    
}

roles = list(data_corrected.keys())
platforms = ['dotwin', 'parler', 'twitter']

# Transpose the data so each role's values for all platforms are together
data_transposed = np.array(list(data_corrected.values())).T

# Position of bars on the x-axis
bar_width = 0.2  # Width of each bar
index = np.arange(len(roles))  # The x locations for the groups

# Create a figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Plot each platform as a group of bars
for i, (platform, color) in enumerate(zip(platforms, colors)):
    plt.bar(index + i * bar_width, data_transposed[i], bar_width, label=platform, color=color)

# Add some text for labels, title, and custom x-axis tick labels, etc.
plt.xlabel('Persistence Levels')
plt.ylabel('Percentage')
plt.xticks(index + bar_width * (len(platforms) - 1) / 2, roles)
plt.ylim(0, 100)
plt.legend(title='Platform',fontsize=14, title_fontsize=16)

# Adding percentage labels on top of each bar
for i, platform in enumerate(platforms):
    for j, v in enumerate(data_transposed[i]):
        plt.text(j + i * bar_width, v + 1, f'{v:.2f}%', ha='center', color='black', fontsize=9)

plt.tight_layout()
plt.show()
```
